```python
import pandas as pd
import os
```

```python
RAW_DATA_PATH = '../data/raw/'

train_file = os.path.join(RAW_DATA_PATH, 'fraudTrain.csv')
test_file = os.path.join(RAW_DATA_PATH, 'fraudTest.csv')
```

```python
df_train = pd.read_csv(train_file)
print("Train shape:", df_train.shape)
print("\nColumns:\n", df_train.columns.tolist())
print("\nFirst 3 rows:\n", df_train.head(3))
```

```python
print("\nData types:\n", df_train.dtypes)
print("\nMissing values:\n", df_train.isnull().sum())
print("\nFraud distribution:\n", df_train['is_fraud'].value_counts(normalize=True) * 100)
```

```python
df_test = pd.read_csv(test_file)
print("Test shape:", df_test.shape)
print("\nFraud distribution in test:\n", df_test['is_fraud'].value_counts(normalize=True) * 100)
```

```python
# Sample for fast prototyping
df_train_sample = df_train.sample(n=10000, random_state=42)
df_train_sample.to_csv('../data/raw/fraudTrain_sample_10k.csv', index=False)
print("Sample saved.")
```
